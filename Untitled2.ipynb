```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.font_manager as fm
from IPython.display import HTML

# 한글 폰트 설정
!apt-get update -qq
!apt-get install fonts-nanum -qq
!fc-cache -fv
path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fontprop = fm.FontProperties(fname=path, size=11)
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# Gravitational acceleration
g = 9.8

def start_animation(x, y):
    # Time taken for the ball to reach the ground
    t_fall = np.sqrt(2 * x / g)
    # Generate time array
    num_frames = 60  # Number of frames in the animation
    t = np.linspace(0, t_fall, num=num_frames)

    # Calculate the trajectory of the ball
    X = y * t
    Y = x - 0.5 * g * t**2

    # Set up the animation
    fig, ax = plt.subplots()
    line, = ax.plot([], [], 'ro', markersize=10)  # Set the color of the ball to red
    path, = ax.plot([], [], 'r-', lw=2)  # Path representing the trajectory
    time_text = ax.text(0.02, 0.95, '', transform=ax.transAxes, fontproperties=fontprop)
    distance_text = ax.text(0.02, 0.90, '', transform=ax.transAxes, fontproperties=fontprop)

    ax.set_xlim(0, X[-1])
    ax.set_ylim(0, x)
    ax.set_xlabel('수평 거리 (m)', fontproperties=fontprop)
    ax.set_ylabel('높이 (m)', fontproperties=fontprop)
    ax.set_title('높이에서 수평으로 던진 공의 궤적', fontproperties=fontprop)

    def init():
        line.set_data([], [])
        path.set_data([], [])
        time_text.set_text('')
        distance_text.set_text('')
        return line, path, time_text, distance_text

    def update(frame):
        line.set_data([X[frame]], [Y[frame]])  # Set each point as an array
        path.set_data(X[:frame+1], Y[:frame+1])
        time_text.set_text(f'걸린 시간 {t[frame]:.2f} 초')
        distance_text.set_text(f'수평 거리 {X[frame]:.2f} m')
        return line, path, time_text, distance_text

    # Set the interval for the animation (displayed in real time)
    interval = t_fall * 1000 / num_frames  # Convert to milliseconds by multiplying by 1000

    ani = animation.FuncAnimation(fig, update, frames=num_frames, init_func=init, repeat=False, interval=interval)

    plt.grid(True)
    return ani

# 높이와 속도 설정
height = 10  # 높이 (m)
speed = 5    # 속도 (m/s)

# 애니메이션 시작
ani = start_animation(height, speed)

# 애니메이션을 HTML로 렌더링
HTML(ani.to_jshtml())
```
